```python
from llama_index.core import VectorStoreIndex,SimpleDirectoryReader,ServiceContext
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core.prompts.prompts import SimpleInputPrompt
```

```python
document = SimpleDirectoryReader(input_dir='./data').load_data()
```

```python
system_prompt = "You are a Q&A assistant. Your role is to deliver accurate answers based on the provided instructions and context. If no context is given, respond with: 'I am an AI trained to assist with reading the red book policies, and I don’t have additional information."
query_wrapper_prompt = SimpleInputPrompt("|USER|>{query_str}<|ASSISTANT|>")
```

```python
import torch

llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",
    device_map="auto",
    # uncomment this if using CUDA to reduce memory usage
    model_kwargs={"torch_dtype": torch.float16 , "load_in_8bit":True}
)
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index.core import ServiceContext
from llama_index.embeddings.langchain import LangchainEmbedding

embed_model=LangchainEmbedding(
    HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2"))
```

```python
from llama_index.core.node_parser import SentenceSplitter
from llama_index.core import Settings

Settings.node_parser = SentenceSplitter(chunk_size=512, chunk_overlap=20)
Settings.llm = llm
Settings.embed_model = embed_model
```

```python
index=VectorStoreIndex.from_documents(document,embed_model=embed_model)
```

```python
query_engine=index.as_query_engine()
```

```python
response=query_engine.query(" Leaves ")
print(response)
```
